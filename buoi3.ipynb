```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

#doc anh
img = cv2.imread("/content/anhanh.jpg",0)

#binarzi the image 
binr = cv2.threshold(img, 0, 255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
#binr = cv2.threshold(img, 0, 255,cv2.THRESH_BINARY)[1]
#binr = cv2.threshold(img, 0, 255,cv2.THRESH_OTSU)[1]


#define the kernel
kernel = np.ones((7,7), np.uint8)

#invert the image 
invert = cv2.bitwise_not(binr)

#erode the image
erosion = cv2.erode(invert, kernel, iterations=1)
```

```python
#print the output
plt.subplot(1,3,1)
plt.title("Oragion Image"), plt.xticks([]), plt.yticks([])
plt.imshow(binr, cmap='gray')
plt.subplot(1,3,2)
plt.title("Invart Image"), plt.xticks([]), plt.yticks([])
plt.imshow(invert, cmap='gray')
plt.subplot(1,3,3)
plt.title("Erode Image"), plt.xticks([]), plt.yticks([])
plt.axis()
plt.imshow(erosion, cmap='gray')
```

```python
#Dilation the image
dilation = cv2.dilate(invert, kernel, iterations=1)

#print the output
plt.subplot(1,3,1)
plt.title("Oragion Image"), plt.xticks([]), plt.yticks([])
plt.imshow(binr, cmap='gray')
plt.subplot(1,3,2)
plt.title("Invart Image"), plt.xticks([]), plt.yticks([])
plt.imshow(invert, cmap='gray')
plt.subplot(1,3,3)
plt.title("Dilation Image"), plt.xticks([]), plt.yticks([])
plt.imshow(dilation, cmap='gray')
```

```python
#Opening operator
#define the kernel
kernel = np.ones((5,5), np.uint8)

#opening the image
opening = cv2.morphologyEx(invert, cv2.MORPH_OPEN, kernel, iterations=1)

#closing operator
closing = cv2.morphologyEx(invert, cv2.MORPH_OPEN, kernel, iterations=1)

#use morph gradient
morph_gradient = cv2.morphologyEx(invert, cv2.MORPH_GRADIENT, kernel)

```

```python
#print the output
plt.subplot(3,2,1)
plt.title("Erosion Image"), plt.xticks([]), plt.yticks([])
plt.imshow(erosion, cmap='gray')
plt.subplot(3,2,2)
plt.title("Dilation Image"), plt.xticks([]), plt.yticks([])
plt.imshow(dilation, cmap='gray')
plt.subplot(3,2,3)
plt.title("Opening Image"), plt.xticks([]), plt.yticks([])
plt.imshow(opening, cmap='gray')
plt.subplot(3,2,4)
plt.title("Closeing Image"), plt.xticks([]), plt.yticks([])
plt.imshow(closing, cmap='gray')
plt.subplot(3,2,5)
plt.title("Morphology Gradient"), plt.xticks([]), plt.yticks([])
plt.imshow(morph_gradient, cmap='gray')

```

```python
import matplotlib
#bai tap 2
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread("/content/anhanh.jpg", cv.IMREAD_GRAYSCALE)
img_blur = cv.GaussianBlur(img,(5,5), sigmaX=0, sigmaY=0)
edges = cv.Canny(img_blur, 100, 200)

plt.subplot(121)
plt.title("Oragion Image"), plt.xticks([]), plt.yticks([])
plt.imshow(img,cmap='gray')
plt.subplot(122)
plt.title("Edge Image"), plt.xticks([]), plt.yticks([])
plt.imshow(edges,cmap='gray')
plt.show()
```
